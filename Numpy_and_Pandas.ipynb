```python
import numpy as np
import pandas as pd
```

```python
#Array Creation

temperatures_w1 = [22.5, 25.3, 20.8, 23.4, 26.1, 24.8, 21.9]
temperatures_w1_array = np.array(temperatures_w1)
print(f"\n Array_1:",temperatures_w1_array)
print(f"\n Shape:",temperatures_w1_array.shape)
print(f"\n Datatype:",temperatures_w1_array.dtype)
print(f"\n Array Size:",temperatures_w1_array.size)

#Array Operations:
print("\n Array Operation")
temperatures_w1_array_fahrenheit = (temperatures_w1_array * 9/5) + 32
print(f"\n temperatures_w1 from Celsius to Fahrenheit:",temperatures_w1_array_fahrenheit)
print(f"\n Maximum fahrenheit",temperatures_w1_array_fahrenheit.max())
print(f"\n Minimum fahrenheit",temperatures_w1_array_fahrenheit.min())
print(f"\n Mean",temperatures_w1_array_fahrenheit.mean())

#Array Slicing and Indexing:

print(f"\n first three days:",temperatures_w1_array_fahrenheit[:3])
print(f"\n last two days:",temperatures_w1_array_fahrenheit[-2:])
print(f"\n middle three days:",temperatures_w1_array_fahrenheit[1:4])

```

```python
#Create a 2D Array

Row1 = [22.5, 25.3, 20.8, 23.4, 26.1, 24.8, 21.9]
Row2 = [19.2, 22.5, 21.3, 24.0, 23.5, 22.8, 20.1]
temperatures = [Row1, Row2]
print(temperatures)


temperatures_array = np.array(temperatures)
print(f"\n temperatures_array:",temperatures_array)
print(f"\n Shape:",temperatures_array.shape)
print(f"\n Datatype",temperatures_array.dtype)
print(f"\n Array Size",temperatures_array.size)
print(f"\n last two days W1:",temperatures_array[0, -2:])
print(f"\n last two days W2:",temperatures_array[1, -2:])
```

```python
#Creating Pandas Series:

marks = pd.Series([95, 92, 89, 85, 80], index=['Rank1','Rank2','Rank3','Rank4','Rank5'])
print(marks)

#Indexing and Slicing:

print(f"\n first rank student:",marks.iloc[0])
print(f"\n 3rd-ranked student:",marks.iloc[2])
print(f"\n The marks are greater than 90:",marks[marks > 90])

#Manipulating Series:

marks.loc['Rank1'] = 100
print(f"\n The marks of the student with the 1st rank:",marks)


Last_rank = marks.index[-1]
marks = marks.drop(Last_rank)
print(f"\n Remove the entry corresponding to the last rank:",marks)

CGPA = marks/10
print(f"\n CGPA:",CGPA)



```

```python
#Pandas DataFrame:

transactions = pd.DataFrame ({'TransactionID':[101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
'ProductCategory':['Electronics', 'Clothing', 'Electronics', 'Furniture', 'Clothing','Electronics', 'Furniture', 'Clothing', 'Furniture', 'Electronics'],
'Region':['North', 'South', 'North', 'East', 'West', 'North', 'East', 'West', 'South','North'],
'Amount':[200, 150, 300, 450, 200, 250, 300, 180, 350, 400]})
transactions

#Data Exploration

print(f"\n Dataframe Head:",transactions.head())
print(f"\n Dataframe Tail:",transactions.tail())
print(f"\n Dataframe Shape:",transactions.shape)
print(f"\n Dataframe Columns:",transactions.columns)
print(f"\n Dataframe Datatype:",transactions.dtypes)
print(transactions[["ProductCategory","Amount"]])
print(f"\n Retrive Last 3 Columns:",transactions.iloc[:,-3:])

filtered_transactions = transactions[
    (transactions['Region'] == 'North') &
    (transactions['Amount'] > 200)
]

print(filtered_transactions)


print(transactions['ProductCategory'].value_counts())

print(transactions['Region'].unique())

mean_amount_by_region = transactions.groupby('Region')['Amount'].mean()

print(mean_amount_by_region)


transactions.loc[transactions['TransactionID'] == 102, 'Amount'] = 165


transactions['Discount'] = transactions['Amount'] * 0.10


transactions = transactions[transactions['TransactionID'] != 109]


transactions = transactions.drop(columns=['Discount'])

print(transactions)
```
